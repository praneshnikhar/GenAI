```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "i was so happy yesterday"

tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)
```

```python
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs:", input_ids)
```

```python
#here it will tokenize also convert to id
encoded_input = tokenizer(text)
print(encoded_input)
```

```python
decoded_input = tokenizer.decode(encoded_input["input_ids"])
print(decoded_input)
```

```python
!pip install arxiv

import arxiv
import pandas as pd
```

```python
query = 'ai OR artificial intelligence OR machine learning OR deep learning'
search = arxiv.Search(query = query, max_results = 10 , sort_by = arxiv.SortCriterion.SubmittedDate)

papers=[]
for result in search.results():
    papers.append({
        'published': result.published,
        'title': result.title,
        'summary': result.summary,
        'abstract': result.categories
    })
df = pd.DataFrame(papers)
pd.set_option('display.max_colwidth', None)
df.head()
```

```python
#summarize any abstract now
abstract = df['abstract'][0]
summarizer = pipeline("summarization", model = "facebook/bart-large-cnn")
summarization_result = summarizer(abstract)
```
